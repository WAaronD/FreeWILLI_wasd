```python
import pandas as pd
import matplotlib.pyplot as plt
import os
import numpy as np
from tracker_log_analysis import *
import glob
```

```python
# Path to the data files
data_dir = "../simulator_program/simulator_data/track103/"
pattern = os.path.join(data_dir, "rawdat_track103_*_DATA.npy")

# Find and sort all matching files
file_list = sorted(glob.glob(pattern))

# Plot first channel from each file
plt.figure(figsize=(15, 6))

for file in file_list:
    data = np.load(file)
    print(data.shape)
    if data.ndim == 2 and data.shape[0] >= 1:
        ch0 = data[:, 0]  # first channel
        plt.plot(ch0, label=os.path.basename(file), alpha=0.7)
        plt.xlabel("Sample Index")
        plt.ylabel("Amplitude")
        plt.show()

#plt.title("First Channel from Each rawdat_track103_*_DATA.npy File")
#plt.xlabel("Sample Index")
#plt.ylabel("Amplitude")
#plt.legend(loc='upper right', fontsize='small', ncol=2)
#plt.tight_layout()
plt.grid(True)
plt.show()
```
